```python
#Replacing all the missing values of 's' and 'na' with the properly designated numpy NaNs
#Created a list of missing values
missing_values = ['s', 'na']
df =pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv', na_values = missing_values)
```

```python
#Dropping columns which are irrelevant/not needed for the data analysis
#Those columns contain the 's', 'na', or very little information which is not
#helpful for the data analysis
drop_cols = ['Number Scoring Below 65', 'Percent Scoring Below 65',
            'Number Scoring 65 or Above', 'Percent Scoring 65 or Above',
            'Number Scoring 80 or Above', 'Percent Scoring 80 or Above',
            'Number Scoring CR', 'Percent Scoring CR']
df.drop(drop_cols, inplace = True, axis=1)
```

```python
#Renaming columns
df.rename(columns={'School Level':'School Grade'})
```

EXAMINATION OF DATA: A. FILTERING / B. COMPARING / C. ANALYZING CHANGES

```python
#Looking for the school DBN for I.S. 141
df[df['School DBN'] == '30Q141']
```

```python
#Creating a new dataframe for I.S. 141 
df2= df[df['School DBN'] == '30Q141']
```

```python
#Confirming the dataframe
df2
```

```python
#Below is the second school that I will be using to compare to I.S.141
#Creating a new dataframe for I.S. 227
df3= df[df['School DBN'] == '30Q227']
```

```python
#Confirming the dataframe
df3
```

```python
#Will combine both of the schools to create a new dataframe
combine = pd.concat([df2, df3], ignore_index=True)
```

```python
#Confirming the combine dataframe
combine
```

```python
#Total of 183 rows and 15 columns
combine.shape
```

```python
#Aggregating Data/Grouping
combine.groupby(['Regents Exam'])['Total Tested'].count()
```

```python
combine.groupby(['Year'])['Total Tested'].count()
```

```python
#Creating a Science dataframe
Science= combine[combine['Regents Exam'] == 'Living Environment']
```

```python
#Confirming the Science dataframe
Science
```
